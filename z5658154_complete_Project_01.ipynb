---

#### **A. Understanding the Health Question and Requirements**

##### **1. Tumour**

Tumours are any lump found in various organs, divided into benign and malignant. Benign means non-cancerous, whereas malignant refers to a cancerous one.

For breast tumours, the benign tumours span a variety of findings, such as
- Fibrocytic changes
- Fibroadenomas
- Papilomas.

Those three are the most common benign breast tumours and the condition can be resolved by removing the cancer by a breast surgeon.

On the other hand, malignant breast tumours are a condition where tumours continuously grow and invade adjacent tissues directly, through the lymphatic system, or bloodstream to different parts of the body. At this stage, physicians need to perform a procedure called a biopsy to detect, assess severity, and determine the treatment strategy. The malignant breast tumours are more difficult to cure.

##### **2. Importance**

Appropriate classification of breast tumour types can benefit patients as well as healthcare services. First, if the tumour is correctly classified as a benign tumour, then we can prevent unnecessary procedures for the patient. Treatment related to benign tumours, for example, for fibroadenoma, is excision rather than going through radiotherapy or chemotherapy. The proper treatment will provide a cost-effective solution that is well-suited to the patient, without incurring unnecessary expenses. Furthermore, the same concept also happens for malignant breast tumours. A correct classification can identify the issue as soon as possible, making the treatment strategy more successful. For malignant breast cancer, physicians need an interprofessional approach to decide which combination of therapy (chemotherapy, radiotherapy, excision) should be used. Early detection of malignant breast cancer in the "localized stage" gives more than 99% 5-year survival compared to the "distant stage" (32%). When classification is done correctly at the GP level, healthcare workers and services can be better prepared for referrals and the necessary treatment and procedures.

   


##### **3. Implication**

For a patient, a false classification can bring harm in both overdiagnosed and underdiagnosed conditions. In an overdiagnosed (false positive) scenario, it will lead to unnecessary testing and even therapy, which can give side effects. Patients can feel anxiety and psychological harm going through the treatment. Additional costs incurred during the procedure can also pose a financial burden for patients if the procedure is not fully covered or if the insurance or government payment scheme does not cover it. For the underdiagnosed breast tumour, a delay in aggressive cancer treatment leads to lethal consequences for the patients.

##### **4. Classification Goal**

##### **5. Method Selection**

Several methods can be used to answer the research question. With output as categorical, from the previous reading materials, the options are as follows.
1. Logistic Regression
 A Linear model specified for classification that needs a regularization
 (l1, l2, or elastic net) to prevent the model from overfitting to the training data.
2. K-nearest-neighbor (KNN) classification
 Classification output to be categorical based on the nearest neighbour classification. The model only consists of storing the training dataset, which later classifies new data points based on the closest data in the training set, precisely like the nearest neighbour (only one or k number of neighbors)
3. Decision Tree Classifier
 Classification using decision boundaries based on Gini impurity or entropy, controlled by the given depth. Tends to overfit to the training dataset; thus, the strategy involves pre-pruning and post-pruning. The maximum depth can also decrease overfitting, but it will result in lower accuracy.
4. Ensemble method
 A series of methods that combine several methods or do one particular method in a specific way to get the final prediction. Options such as Voting classifier, bagging (pasting) classifier, random forest classifier, gradient boosted classifier, and stacking classifier are available.

Regarding the research question in this project, **logistic regression** can serve as a starting point because the nature of the linear model is straightforward to understand and interpret. This method is robust to small datasets and suited for binary classification. In a medical context, identifying the critical features for the model can be achieved by using regularization in logistic regression, which helps highlight features that require specific observation, particularly in evaluating tumor malignancy. This hypothesis needs to be verified using **GridSearchCV** to determine the optimal combination of hyperparameters.

##### **6. Method Evaluation - Success Criteria**

Evaluation metrics to best capture "success" in the medical context for breast tumour is to get the most correct diagnosis of malignant breast tumor by making sure the false negative rate does not occur widely. Having results that show patients as healthy instead of accurately reflecting their actual malignant condition causes considerable harm. #### 9. Train-Test Split (Grouped by ID and Stratified by Diagnosis)

To get the preparation for model, train-test split need to be done. We perform train-test split by `ID` with stratification based on `diagnosis`. This steps ensures that:
 - Each `ID` only appears in one partition (no leakage)
 - Class proportions maintained across split

Since modeling should exclude patient identifiers, we construct `df_clean` by dropping the `ID` column from `df_intermediate`. This ensures no accidental leakage during training.

The split based on df_intermediate that still have `ID` within dataset. However for the model features (`X`) and labels (`y`) taken from `df_clean` where `ID` is removed.

The detaild steps are as follow
1. Defining `df_clean` for `X` and `y` references.
2. Taking unique `ID` and `diagnosis` label within the `ID` and assign to `id_labels`.
3. Splitting using `id_labels['ID']`as input and stratifying on `id_labels['diagnosis']`.
4. Building mask for train (`train_mask`) and test (`test_mask`) taken from `df_intermediate` that still has `ID`.
5. Defining `X` and `y` from `df_clean`.
6. Applying masks to split modelling data.

***References***
1. The Machine Learning Landscape in Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow, 3rd Edition. (n.d.). Retrieved July 10, 2025, from https://learning.oreilly.com/library/view/hands-on-machine-learning/9781098125967/ch01.html
1. https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
2. https://copilot.microsoft.com/shares/pages/1HTyZhDSz6qnGVYx5heDH
3. https://scikit-learn.org/stable/common_pitfalls.html

```python
# Step 1: Defining `df_clean`
df_clean = df_intermediate.copy().drop(columns='ID')
```

```python
# Step 2: Mapping unique `ID` to its diagnosis label
id_labels = df_intermediate[['ID', 'diagnosis']].drop_duplicates()

# Step 3: Performing stratified splid based on diagnosis labels
id_train, id_test = train_test_split(
    id_labels['ID'],
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=id_labels['diagnosis']
)

# Step 4: Building train - test masks using `df_intermediate (still has `ID)
train_mask = df_intermediate['ID'].isin(id_train)
test_mask  = df_intermediate['ID'].isin(id_test)

# Step 5: Preparing `X` and `y` from `df_clean`
X = df_clean.drop(columns='diagnosis')
y = df_clean['diagnosis']

# Step 6: Applying masks to split the modelling data
X_train, X_test, y_train, y_test = X[train_mask], X[test_mask], y[train_mask], y[test_mask]
```

## <font color="green">**Question 3: Training and Tuning**

---

In this section, we train and tune hyperparameters of a LogisticRegression model.


### Step 1: Defining Logistic regression and pipeline




First, we define the Logistic Regression model, that later can be used inside the pipeline and passed to `GridSearchCV`. The solver set to `liblinear` as this algorithm supports `l1` and `l2` `penalty.

The random_state and n_jobs taken from the constant RANDOM_STATE (2025) and N_JOBS (1) that already defined.

Next, we define the pipeline for assemble several steps that can be cross-validated together. Here we put `scaler` using `StandardScaler` and for `classifier` taking the defined `log_reg`

```python
# Defining pipeline
pipeline = Pipeline([('scaler', StandardScaler()), ('classifier', log_reg)])
```

### Step 2: Tuning three hyperparameters: `penalty`, `C1`, and `class_weight`

For `GridSearchCV`, we need to define the `param_grid`. Several parameter as stated need to be test
1. `penalty` testing `l1` and `l2`
2. `C` testing for 0.05, 0.5, 5, 50, and 500
3. `class_weight`, here we decide to weighted in Class 1 with more weight rather than class 0. The reason behind this choice is that class 1 represent malignant which in the datase is imbalance (more benign), thus we choose **Option A**

 - (80% for class 1, 20% for class 0)
 - (70% for class 1, 30% for class 0)
 - (60% for class 1, 40% for class 0)
 - None (i.e., no weights)

***References***
1. compute_class_weight. (n.d.). In https://scikit-learn.org/stable/modules/generated/sklearn.utils.class_weight.compute_class_weight.html.
2. Glossary of Common Terms and API Elements. (n.d.). https://Scikit-Learn.Org/Stable/Glossary.html#term-Class_weight.

```python
# Defining param_grid
param_grid = {'classifier__penalty': ['l1', 'l2'],'classifier__C': [0.05, 0.5, 5, 50, 500],'classifier__class_weight': [{1: 0.8, 0: 0.2}, {1: 0.7, 0: 0.3},{1: 0.6, 0: 0.4},None]}
```

### Step 3: Incorporating `GridSearchCV` using `StratifiedGroupKFold` and `f1` scoring for evaluation


For the cross validation, we need to ensure there are no data leakage from the row that have same ID. To do so, we use `StratifiedGroupKFold` instead of default `cv` that use regular `KFold`.

The `StratifiedGroupKFold` approach ensures
- Each fold contains a similar distribution of diagnosis classes (stratification)
- All samples from a given patient (ID) are kept within a single fold (grouping), preventing leakage across folds, especially important when multiple rows exist for a single patient.

This approach is essential in clinical modelling where in real life there can be repeated measures from same individuals.

The number partitions that used is 4 partition (k=4) as defined in `N_SPLIT` constant that gives **3-fold cross-validation**.

For the scoring metrics, `F1-score` picked because it balanced precision and recall, appropriate for detecting malignant cases where false negative are clinically critical.

Number of jobs set from constant `N_JOBS` that already define (set as 1) that actualy the default to constraint parallel computation.

The remaining parameters for `GridSearchCV` remains unchanged.

***References***
1. GridSearchCV. (n.d.). https://Scikit-Learn.Org/Stable/Modules/Generated/Sklearn.Model_selection.GridSearchCV.html.


```python
gs_log_reg = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv= StratifiedGroupKFold(N_SPLITS), scoring='f1', n_jobs=N_JOBS)
```

### Step 4: Fitting Grid Search to Training Set and use to Evaluate Test Set

Now running the `GridSearchCV` to the training set, to make it work allign with the cross-validation approach, we give details of parameter`groups` in `gs_log_reg.fit()` by using group by `ID` taken from `train_mask`

```python
# Fitting the grid search on the training data
gs_log_reg.fit(X_train, y_train, groups=df_intermediate.loc[train_mask, 'ID'])
```

We can check the best model and the performance to know what happen behind the scene.

```python
# Checking best model and the performance
print("Best Model Parameters:", gs_log_reg.best_params_, "Best F1 Score:", gs_log_reg.best_score_)
```

The best parameters shows that `penalty` is `l2`, `C` is 0.5, and `class_weight` 60% to class 1 and 40% class 0. From the finding we can evaluate models on both training and test set using `classification report` and `confusion matrix`. The evaluation put inside a function to ensure the code can be used by training and tes set without the need to write for each.

```python
def evaluate_model(y_true, y_pred, set_name="Set"):
  """Evaluating classification performance and visualises confusion metric

  Args:  y_true : array-like (true label for dataset, e.g. training/ test)
         y_pred : array like, predicted label from the model
         set_name : Name dataset being evaluated
  Retunrs:
  None
  """

  print(f"Classification Report on {set_name}:")
  print(classification_report(y_true, y_pred, digits=3))

  cm = confusion_matrix(y_true, y_pred)
  class_labels = ["Benign", "Malignant"]

  plt.figure(figsize=(6, 5))
  sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
  plt.title(f"{set_name} Confusion Matrix")
  plt.xlabel("Predicted Label")
  plt.ylabel("True Label")
  plt.tight_layout()
  plt.show()
```

```python
# Training set
evaluate_model(y_train, gs_log_reg.predict(X_train), "Training Set")
```

```python
# Test set
evaluate_model(y_test, gs_log_reg.predict(X_test), "Test Set")
```
